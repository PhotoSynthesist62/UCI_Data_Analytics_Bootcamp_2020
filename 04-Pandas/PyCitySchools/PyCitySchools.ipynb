## PyCitySchools

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = 'Resources/schools_complete.csv'
student_data_to_load = 'Resources/students_complete.csv'

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load,encoding="utf-8")
student_data = pd.read_csv(student_data_to_load,encoding="utf-8")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how='left', on=['school_name', 'school_name'])
school_data_complete.head()
```

### District Summary
* Total number of schools
* Total number of students
* Total budget
* Ave math score 
* Ave reading score
* Percentage of students with a passing math score (70 or greater)
* Percentage of students with a passing reading score (70 or greater)
* Percentage of students who passed math **and** reading (% Overall Passing)
* Create dataframe to hold the above results
* Optional: give displayed data cleaner formatting

```python
# 01. Ct unique 'school_name' = 'total_schools'; add 'total_schools' to 'dist_summ_df' 
total_schools = len(school_data_complete['school_name'].unique())

# 02 Ct unique 'Student ID' = 'total_students'; add 'total_students' to 'dist_summ_df'
total_students = len(school_data_complete['Student ID'].unique())

# 03. Sum 'budget' = 'total_budget', add 'total_budget' to 'dist_summ_df' 
total_budget = sum(school_data_complete['budget'].unique())

# 04. Ave 'math_score' = 'ave_math'; add 'ave_math' to 'dist_summ_df' 
ave_math = school_data_complete['math_score'].mean()

# 05. Ave 'read_score' = 'ave_read'; add 'ave_read' to 'dist_summ_df' 
ave_read = school_data_complete['reading_score'].mean() 

# 06. Create series of students passing math >=70; cacl percentage
m_pass_dex = school_data_complete.loc[school_data_complete['math_score'] >= 70]
m_pass_pct = (len(m_pass_dex))/total_students

# 07. Create series of students passing reading >=70; cacl percentage
r_pass_dex = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
r_pass_pct = (len(r_pass_dex))/total_students

# 08. Create series of students passing reading & math; cacl percentage >=70
o_pass_dex = m_pass_dex + r_pass_dex
o_pass = o_pass_dex.drop_duplicates(subset='Student ID')
o_pass_pct = (len(o_pass)) / total_students

# 09. Create a new dataframe for District Summary holding: 
              
dist_summ_df = pd.DataFrame({'Total No. of Schools': [total_schools],
                            'Total No. of Students': [total_students],
                            'Total Budget': [total_budget],
                            'Ave Math Score': [ave_math],
                            'Ave Reading Score': [ave_read],
                            '% Passing Math': [m_pass_pct],
                            '% Passing Reading': [r_pass_pct],
                            '% Passing Overall': [o_pass_pct]})

# 10. Format for Presentation
dist_summ_df['Total No. of Schools'] = dist_summ_df['Total No. of Schools'].map('{:,}'.format)
dist_summ_df['Total No. of Students'] = dist_summ_df['Total No. of Students'].map('{:,}'.format)
dist_summ_df['Total Budget'] = dist_summ_df['Total Budget'].map('{:,.0f}'.format)
dist_summ_df['Ave Math Score'] = dist_summ_df['Ave Math Score'].map('{:,.2f}'.format)
dist_summ_df['Ave Reading Score'] = dist_summ_df['Ave Reading Score'].map('{:,.2f}'.format)
dist_summ_df['% Passing Math'] = dist_summ_df['% Passing Math'].map('{:,.2f}'.format)
dist_summ_df['% Passing Reading'] = dist_summ_df['% Passing Reading'].map('{:.2f}'.format)
dist_summ_df['% Passing Overall'] = dist_summ_df['% Passing Overall'].map('{:,.2f}'.format)
dist_summ_df
```

### School Summary
Create an overview table that summarizes key metrics about each school, including:
* School Name
* School Type
* Total Students
* Total School Budget
* Per Student Budget
* Average Math Score
* Average Reading Score
* % Passing Math
* % Passing Reading
* % Overall Passing (The percentage of students that passed math **and** reading.)
* Create a dataframe to hold the above results

```python
# 01. Group by 'school_name'
school_gp = school_data_complete.groupby(['school_name'])
s_type = school_data.set_index(['school_name'])['type']
s_students = school_data_complete['school_name'].value_counts()

# 02. Create series for School Budget, Ave Math Score, Ave Reading Score
s_budget = school_gp.mean()['budget']

s_stud_budget = (s_budget / s_students)

#03. Create series of students for Average Math Scores
s_math_ave = school_gp.mean()['math_score']

#04. Create series of students for Average Reading Scores
s_reading_ave = school_gp.mean()['reading_score']

# 05. Create series of students passing math >=70 ; cacl percentage 
s_m_pass_dex = school_data_complete.loc[school_data_complete['math_score'] >= 70]
s_m_pass = s_m_pass_dex['school_name'].value_counts()
s_m_pass_pct = s_m_pass / s_students

# 06. Create series of students passing reading >=70 ; cacl percentage 
s_r_pass_dex = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
s_r_pass = s_r_pass_dex['school_name'].value_counts()
s_r_pass_pct = s_r_pass / s_students

# 07. Create series of students passing BOTH math and reading >=70 ; cacl percentage
s_o_pass_dex = school_data_complete.loc[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
s_o_pass = s_o_pass_dex['school_name'].value_counts()
s_o_pass_pct = s_o_pass / s_students

# 08. Combine series into a new dataframe for Schools Summary holding: 
school_summary = pd.DataFrame({'School Type': s_type,
                                'No. of Students': s_students,
                                'School Budget': s_budget,
                                'Budget per Student': s_stud_budget,
                                'Ave Math Score': s_math_ave,
                                'Ave Reading Score': s_reading_ave,
                                '% Passing Math': s_m_pass_pct,
                                '% Passing Reading': s_r_pass_pct,
                                '% Passing Overall': s_o_pass_pct
                              })

# 09. Format for Presentation - formatting breaks subsequent code
#school_summary['No. of Students'] = school_summary['No. of Students'].map('{:,}'.format)
#school_summary['School Budget'] = school_summary['School Budget'].map('{:,.0f}'.format)
#school_summary['Budget per Student'] = school_summary['Budget per Student'].map('{:,.0f}'.format)
#school_summary['Ave Math Score'] = school_summary['Ave Math Score'].map('{:,.2f}'.format)
#school_summary['Ave Reading Score'] = school_summary['Ave Reading Score'].map('{:,.2f}'.format)
#school_summary['% Passing Math'] = school_summary['% Passing Math'].map('{:,.2f}'.format)
#school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map('{:.2f}'.format)
#school_summary['% Passing Overall'] = school_summary['% Passing Overall'].map('{:,.2f}'.format)
school_summary
```

### Math Scores by Grade
Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
* Create a pandas series for each grade. Hint: use a conditional statement.
* Group each series by school
* Combine the series into a dataframe
* Optional: give the displayed data cleaner formatting

```python
# 01. Create a pandas series for each grade. Hint: use a conditional statement.
ninth_s = school_data_complete.loc[school_data_complete['grade'] == '9th',:]
tenth_s = school_data_complete.loc[school_data_complete['grade'] == '10th',:]
eleventh_s = school_data_complete.loc[school_data_complete['grade'] == '11th',:]
twelfth_s = school_data_complete.loc[school_data_complete['grade'] == '12th',:]

# 02. Group each series by school
ninth_gb = ninth_s.groupby('school_name')
tenth_gb = tenth_s.groupby('school_name')
eleventh_gb = eleventh_s.groupby('school_name')
twelfth_gb = twelfth_s.groupby('school_name')

# 03. Create series of students for Average Math Scores
ave_read_9 = ninth_gb['math_score'].mean()
ave_read_10 = tenth_gb['math_score'].mean()
ave_read_11 = eleventh_gb['math_score'].mean()
ave_read_12 = twelfth_gb['math_score'].mean()

# 04. Combine the series into a dataframe
grade_summary = pd.DataFrame({
                '9th Grade Math Score': ave_read_9,
                '10th Grade Math Score': ave_read_10,
                '11th Grade Math Score': ave_read_11,
                '12th Grade Math Score': ave_read_12
        })

# 06. Make it Pretty
grade_summary['9th Grade Math Score'] = grade_summary['9th Grade Math Score'].map('{:,.2f}'.format)
grade_summary['10th Grade Math Score'] = grade_summary['10th Grade Math Score'].map('{:,.2f}'.format)
grade_summary['11th Grade Math Score'] = grade_summary['11th Grade Math Score'].map('{:,.2f}'.format)
grade_summary['12th Grade Math Score'] = grade_summary['12th Grade Math Score'].map('{:,.2f}'.format)
grade_summary
```

### Scores by School Spending
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# 01. Find Max and Min
#print(school_data['size'].max())
#print(school_data['size'].min())
#print(school_data['size'].mean())

# 02. Create 4 bins
budget_bins = [0, 600, 625, 650, 675]

# 02. Label the bins
budget_labels = ['$ < $600k', '$ 600 - 625k', '$ 625 - 650k', '$ 650 - 675k']

# 03. Slice data for the bins 
pd.cut(school_summary['Budget per Student'], budget_bins, labels=budget_labels)
school_summary['Spending Category'] = pd.cut(school_summary['Budget per Student'], budget_bins, labels=budget_labels)
spend_group = school_summary.groupby('Spending Category')
school_spend_summary = spend_group[['Ave Math Score','Ave Reading Score','% Passing Math','% Passing Reading']].mean()
# Overall Passing Rate (Average of the above two)
school_spend_summary
```
